## подсчет метрик TPR, FPR «вручную»

```python
def find_matrix_error(predictions, y_test, treshold):
    tp = 0 # True positive
    fp = 0 # False positive
    fn = 0 # False negative
    tn = 0 # True negative
    for predicted_prob, actual in zip( predictions[:, 1], y_test ):    
        if predicted_prob >= treshold:
            predicted = 1
        else:
            predicted = 0
        
        if predicted == 1:
            if actual == 1:
                tp += 1
            else:
                fp += 1
        
        else:
            if actual == 1:
                fn += 1
            else:
                tn += 1
    return tp, fp, fn, tn
            
```

```python
# Если честно, так и не понял какие брать значения. Один эксперт написал что можно взять от 0 до 1 с шагом 0.1
# другой пишет что этого недостаточно.
# Взял шаг 0.01 
treshold = np.arange(0, 1.01, 0.01)
fpr_2 = []
tpr_2 = []
for treshold_item in treshold:
    tp_item, fp_item, fn_item, tn_item = find_matrix_error(predictions, y_test, treshold_item)
    fpr_item = fp_item / (fp_item + tn_item)
    tpr_item = tp_item / (tp_item + fn_item)
    fpr_2.append(fpr_item)
    tpr_2.append(tpr_item)
```

```python
plt.plot( fpr_2, tpr_2 )
plt.title('ROC-кривая, посчитанная вручную')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.show()
```

## Сравниваем результаты 2 вариантов рассчета ROC-кривой

```python
#рисуем диагональную линию от 0 до 1
diag_line = [0,1]

plt.plot( fpr, tpr, label='посчитана автоматически')
plt.plot( fpr_2, tpr_2, label='посчитана вручную' )
plt.plot(diag_line, label="диагональная линия")
plt.title('Сравниваем результаты')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.legend()
plt.show()
```

## График Precision-Recall

```python
precision = []
recall = []
for treshold_item in treshold:
    tp_item, fp_item, fn_item, tn_item = find_matrix_error(predictions, y_test, treshold_item)
    precision_denominator = tp_item + fp_item
    # в ответах на вопросы написали, что при делении на ноль показатель можно занулить
    # для построения красивого графика пропускаю такие значения
    if(precision_denominator == 0):
        continue
    else:
        precision_item = tp_item / (tp_item + fp_item)
    recall_item = tp_item / (tp_item + fn_item)
    precision.append(precision_item)
    recall.append(recall_item)
```

```python
plt.plot( recall, precision)
plt.title('Кривая Precison-Recall')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.show()
```

# Выводы по проделанной работе

В идеальном случае, ROC-кривая будет проходить от координат (0,0) строго вверх до (0,1), затем строго направо до коориднаты (1,1). Чем ближе наш график к идеальному, тем лучше модель справляется со своей задачей.

ROC-кривая в данном случае далека от диагонали, что говорит о том, что это не случайное угадывание + пик относительно близок к координате (0,1) что говорит нам о том, что модель хорошо справляется со своей задачей.
